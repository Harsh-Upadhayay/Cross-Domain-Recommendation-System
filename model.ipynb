```python
# Import dependencies
import pandas as pd
import numpy as np

# Load the dataset in a dataframe object and include only four features as mentioned
url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/train.csv"
df = pd.read_csv(url)
include = ['Age', 'Sex', 'Embarked', 'Survived'] # Only four features
df_ = df[include]
df_
```

```python
for col in df.columns:
    print(df[col].dtype.name)

df = df_.dropna()

df
```

```python
df_ohe = pd.get_dummies(df, columns=['Sex', 'Embarked'])

df_ohe
```

```python
from sklearn.linear_model import LogisticRegression

dependent_variable = 'Survived'
x = df_ohe[df_ohe.columns.difference([dependent_variable])]
y = df_ohe[dependent_variable]

print(x, y)
```

```python

lr = LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=1000000000, multi_class='ovr', n_jobs=1,
          penalty='l2', random_state=None, solver='liblinear', tol=0.00000000000000000000000000000001,
          verbose=0, warm_start=False)

lr.fit(x, y)

print(lr)
```

```python
lr.predict([[22, False, False, True, False]])
```
